    *  Because of the min and max speed differences it shows we should **eliminate outliers** with high variance between the speed columns
    * This will drop extreme differences in batted ball reads between System A and System B

```python
def remove_outliers_IQR(df, col):
    
    q1, q3 = np.percentile(df[col], [25, 75])
    print("q1: ", q1)
    print("q3: ", q3)

    IQR = q3-q1
    print("IQR: ", IQR)

    upper_bound = q3 + 1.5 * IQR
    print("upper_bound: ", upper_bound)
    lower_bound = q1 - 1.5 * IQR
    print("lower_bound: ", lower_bound)

    df = df.loc[(df[col] < upper_bound) & (df[col] > lower_bound)] 
    
    return df

battedBall_df = remove_outliers_IQR(battedBall_df, "speedDiff")
```

* #### This function uses interquartile range as a baseline for eliminating outliers in the speedDiff column

```python
print("shape: ", battedBall_df.shape)

battedBall_df["avgSpeed"] = (battedBall_df['speed_A'] + battedBall_df['speed_B']) / 2
battedBall_df["avgAngle"] = (battedBall_df['vangle_A'] + battedBall_df['vangle_B']) / 2

features = ["speed_A", "vangle_A", "speed_B", "vangle_B", "speedDiff", "angleDiff", 'avgSpeed', 'avgAngle']
```

```python
#i have a normal distribution on the means
meanDiff = battedBall_df.speedDiff.mean().round(3) #d
print("Mean Speed Diff: ", meanDiff)
stdDiff = battedBall_df.speedDiff.std().round(3) #s
print("STD of Diff: ", stdDiff)

l_bound = (meanDiff - 1.96*stdDiff).round(3)
u_bound = (meanDiff + 1.96*stdDiff).round(3)
print("l bound: ", l_bound.round(3)) #limit of agreement L
print("u bound: ", u_bound.round(3)) #limit of agreement U
#standard error of meanDiff
std_error = stdDiff / np.sqrt(len(battedBall_df.speed_A))
std_error = np.sqrt(stdDiff*2 / len(battedBall_df.speed_A))
print("std error: ", std_error)
```

## Post Outlier Removal Statistical Analysis

I wanted to evaluate whether the two systems violated or accepted the limits of agreement theory. The limits of agreement theory compares whether or not two different measurement systems, System A and System B, have an "agreement" in how they are measuring the same data. As long as 95% of the speed differences falls between the upper and lower limits range I can assume that the two systems accept this term of agreement.

In comparison, System B has more volatility in its overall exit velocity reads, shown by a higher standard deviation. We can assume that if we eliminate extreme reads from System A and System B, we can use a regression model to estimate the true exit velocity of a ball.

## Findings

I was able to discover that through summary tables and visualizations, each System has its own quirks based in the variance of exit velocity and launch angle reads. Before any cleaning, I was able to assess that the launch angle readings from each System were fairly consistent between them. System A and System B each had higher variance with extremely negative reads and reads above the 90th percentile. This shows that in initial readings, each system struggles to accurately measure extreme launch angles. More so, this same trend applied to exit velocity reads as the percentile ranks between the two systems is on very different scales. I addressed how I was able to analyze that System A was juiced in readings in comparison to System B. After coming to this conclusion, I realized that there needs to be a happy medium between the two systems. System A has higher exit velocity reads, while System B has less variance in its launch angle reads, so I decided to explicitly compare how different the two metrics were for each batted ball read between the two systems.

## Analysis Plan

My analysis plan wasn't able to be fully executed due to time constraints, but I will outline what my attack plan is. I wanted to normalize the mean differences, so I used the IQR strategy to eliminate extreme reads in the dataset. An extreme read is classified by a high disparity difference in batted ball measurements between the two systems. The extreme reads would affect how we compare the quirks and differences between the two systems. I then wanted to analyze the average reads of each batted ball between the two systems to use as a baseline for "true" exit velocity reads. I went along with the idea that if I can clean out misreads on both systems and use them together to get an average read for each batted ball, I can nullify each system's quirks through intense data cleaning and thus making them more normalized. 

To accommodate for batted balls that are missing from either System A or System B, I plan on creating functions that each use a linear regression model to project the exit velocity of the other system as a project of the other system's launch angle, exit velocity, and projected system's launch angle. For example, if I am projecting a batted ball's exit velocity on System A's specs while only having System B's data, I would use System B's launch angle and exit velocity of the batted ball. However, because I don't know the launch angle of the batted ball on System A's specs, I would adjust System B's launch angle to estimate System A's read on the batted ball's launch angle. Specifically, I would add the mean difference in launch angle between the two Systems to System B's launch angle measurement and declare it as System A's "expected" launch angle. By using this general assumption of launch angle for System A, we can use it as a predictor for estimating what the batted ball's exit velocity would be on System A. The inverse of this strategy would be applied to a batted ball whose data was only known from a System A measurement.

After finding the batted ball's exit velocity on both systems, we can average them out and use that as our projected true exit velocity of a batted ball. After projecting the "xEV" for each batted ball by using these methods, we can aggregate our findings by using a group by function on the hitters in the dataset to assess what hitters are boding well for future development or what hitters are regressing based on their past data. The aggregated hitter data of average "xEV" for their batted balls can be used to funnel up to higher level models to further evaluate their true talent as a hitter regardless of what tracking system is used.

## Conclusion

My goal of the analysis bore more focus on the intricacies of each system rather than projecting future exit velocities as a product of only launch angle and the batted ball type. Projecting exit velocity without barrel and bat speed metrics is futile in nature because the launch angle only measures the direction and fails to account for the impact of the bat on the baseball. I was able to discover through statistical analysis how the two systems work with their own quirks to measure the same batted ball event and thus was able to mitigate the effects of their quirks to create a more "normalized" measurement of a batted ball based on the Gaussian Distribution of mean differences. In theory, we can project a "true" measurement if we can reduce the influence of unique outside variables between two different measurement systems. The limits of agreement theory acknowledge this and give light on how to attack a problem of this nature.
